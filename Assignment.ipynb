```python
# calulate the area of circle
#The formula for Area of circle is πr**2
#r = radius

from math import pi as π
r = float(input ('Radius (cm):  '))
Area = (π*(r**2))
Area = str(round(Area,2))
print ("The Area of a Circle is  "  + Area)
```

```python
# the formula for simple intereat is I= (PTR)/100
'''P = principal, T = time, R = rate'''
print ('calculating simple interest')
P = float(input('principal:  '))
T = int(input('time:  '))
R = float(input('rate:  '))
I = (P*T*R)
I = str(round(I,2))
print ('The simple interest of principal '+'N'+ str(P) + 'k for ' + str(T) + 'years, at rate ' + str(R) + '% is equals to ' + 'N' + I +'k')
```

```python
#My Assays Schedule

while True:
    name = input('Welcome \nPlease input your name:  ')
    Reagents_buffers = [['DNTB','Phosphate Buffer','CDNB','GSH','H2O2'],['Buffer C','Buffer D','MSH','Rotenone','Succinate'],
                    ['TBA','TCA','Tris-KCl']]
    Assays = ['Mitochondrial','Anti-oxidant','LPO']
    Assay = input('Goodday  '+ name + '\nWhat Assay do you want to do?  \nMitochondrial, Gene expression, Anti-oxidant or LPO: ')
    
   
    #End = break
    
    if Assay == Assays[0]:
        print('These are what you need ' + str({Assay:Reagents_buffers[1]}))
    
    elif Assay == Assays[1]:
        print('These are what you need ' + str({Assay:Reagents_buffers[0]}))

    elif Assay == Assays[2]:
        print('These are what you need ' + str({Assay:Reagents_buffers[2]}))
    
    elif Assay != Assays:
        print('\nPlease see the laboratory techonologist for the reagents you need')
    
    print ('---------------------------------------------------------------------------')
        
    Continue = input('\nEnter "Continue" to choose another assay or "End" to quit:   ')   
    
    if Continue == "Continue":
        print(Assay)
        
    else:
        break
```
